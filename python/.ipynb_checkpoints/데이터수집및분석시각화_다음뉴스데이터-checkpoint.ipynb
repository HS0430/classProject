```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

import requests
import bs4

url = 'https://news.daum.net/'
res = requests.get(url)

# print(res.text)

soup = bs4.BeautifulSoup(res.text)
data = soup.find_all('a')
# print(data)
font_path = '/Windows/Fonts/batang.ttc'
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

category_url = {
    '메인' : ['https://news.daum.net/'],
    '사회' : ['https://news.daum.net/society'],
    '정치' : ['https://news.daum.net/politics'],
    '경제' : ['https://news.daum.net/economic'],
    '국제' : ['https://news.daum.net/foreign'],
    '문화' : ['https://news.daum.net/culture'],
    'IT' : ['https://news.daum.net/digital'],
}

for i, j in category_url.items() :
    res = requests.get(j[0])
    j.append(res.text)
```

```python
for i, j in category_url.items() :
    soup = bs4.BeautifulSoup(j[1])
    # 언론사 리스트
    temp = []
    
    for n in soup.find_all('span', attrs={'class':'info_cp'}) :
        temp.append(n.text.strip())
        
    for n in soup.find_all('span', attrs={'class':'info_thumb'}) :
        if n.find('a') == None :
            if '\n' in n.text.strip() :
                temp.append(n.text.strip().split('\n')[0])
            elif n.text.strip() == '' :
                temp.append(n.find('img', alt=True)['alt'])
            else :
                temp.append(n.text.strip())
        
    j.append(temp)
```

```python
# 모든 언론사 리스트
data = []

for i, j in category_url.items() :
    if i != '메인' :
        data.extend(j[2])

unique_data = list(set(data))
```

```python
# 분야별 언론사 랭킹
rank = {}
part = ['사회', '정치', '경제', '국제', '문화', 'IT']

plt.figure(figsize=(10, 10))

p = part[1]

for i in unique_data :
    rank[i]=1

for i in category_url[p][2] :
    rank[i] += 1
    
x = rank.keys()
y = rank.values()

plt.title(p+' 분야 언론사 랭킹')
plt.bar(x, y, color='lightgray', edgecolor='black', linewidth=1, width=0.7)
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))

for i in category_url.keys() :
    rank = {}
    
    for j in unique_data :
        rank[j] = 0
    
    for j in category_url[i][2] :
        if i == '메인' :
            continue
        rank[j] += 1
        
    x = rank.keys()
    y = rank.values()
    plt.bar(x, y, label=i)
    
plt.xticks(rotation=90)
plt.legend()
plt.title('전체 분야별 언론사 랭킹')
plt.show()
```

```python
# 전체 랭킹
rank2 = {}

plt.figure(figsize=(10, 10))

for i in data:
    if rank2.get(i) == None :
        rank2[i] = 1
    else:
        rank2[i] += 1

x2 = rank2.keys()
y2 = rank2.values()

plt.bar(x2, y2, color='lightgray', edgecolor='black', linewidth=1, width=0.7)
plt.xticks(rotation = 90)
plt.title('전체 분야 언론사 랭킹')
plt.show()
```
