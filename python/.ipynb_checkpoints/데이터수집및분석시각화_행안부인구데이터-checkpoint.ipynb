```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_path = '/Windows/Fonts/batang.ttc'
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

plt.figure(figsize=(10, 10))

result = pd.read_csv('201312.csv', encoding='cp949')

gu = {}
j = 55 # 연도별 컬럼 값 +6 단위로 변경

for i in range(1, len(result)) :
    gu[i] = result.iloc[i, 0].replace('서울특별시', '').split('(')[0].strip()
    result.iloc[i, j] = int(result.iloc[i, j].replace(',', ''))
    plt.bar(gu[i], result.iloc[i, j], width=0.7, color='lightgray', edgecolor='black', linewidth=1)
    
plt.title(result.columns[j].split('년_')[0].strip()+'_구별 총 인구수')
plt.xticks(rotation=90)
plt.show()
```
